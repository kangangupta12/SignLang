```python
import os
import random
import matplotlib.pyplot as plt
from src.data import list_samples

DATASET_DIR = "dataset/videos"
USE_FRAMES = False

paths, labels = list_samples(DATASET_DIR, USE_FRAMES)
print(f"Total samples: {len(paths)}")
```

```python
from collections import Counter
import pandas as pd

counts = Counter(labels)
df = pd.DataFrame(counts.items(), columns=["class", "count"]).sort_values("count", ascending=False)
df.plot(kind="bar", x="class", y="count", figsize=(12,4), title="Class distribution")
```
